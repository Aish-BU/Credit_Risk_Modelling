## Introduction

Credit risk modeling is a crucial component for financial institutions to effectively manage and mitigate the risk of borrower default on loans, credit cards, or other types of credit products. This risk arises when borrowers are unable to repay the full amount owed, including principal and interest, sometimes making only partial payments. To address this, financial institutions leverage both statistical and machine learning techniques to build robust predictive models that assess creditworthiness, manage exposure, and comply with regulatory standards.

### Key Concepts and Formulas

In credit risk management, the Internal Ratings-Based (IRB) approach is widely used, with two main variants:

* **Foundation IRB:** Where Probability of Default (PD) is internally estimated by the bank, while Loss Given Default (LGD) and Exposure at Default (EAD) values are prescribed by regulators.
* **Advanced IRB:** Where PD, LGD, and EAD are all estimated internally by the bank using advanced models.

Key definitions are:

* **Probability of Default (PD):** The likelihood that a borrower will default on their loan within a one-year horizon, often modeled using logistic regression. It is expressed as a percentage between 0% and 100%, indicating the borrower's risk level.

* **Loss Given Default (LGD):** The proportion of the exposure that is lost if a borrower defaults. It is calculated as $LGD = 1 - \text{Recovery Rate}$ and is commonly modeled using beta regression.

* **Exposure at Default (EAD):** The total value the bank is exposed to when a borrower defaults. It estimates the outstanding amount at the time of default and is also modeled with beta or linear regression techniques.

* **Expected Loss (EL):** The anticipated average loss due to default, computed as:

$$
\text{EL} = \text{PD} \times \text{LGD} \times \text{EAD}
$$

This formula integrates the three core risk components into a single metric used for capital allocation, pricing, and risk management.

---

### Types of Credit Risk Models

Credit risk modeling can be divided broadly into two types:

1. **Application Models:**

   * Used to decide whether to grant a loan and at what interest rate.
   * Employ risk-based pricing where higher risk leads to higher interest rates.

2. **Behavioral Models:**

   * Assess whether to extend additional credit to existing borrowers.
   * Often implemented via scorecards derived from PD models.

#4 Discrete Data Preparation - WoE

**Let's automate the calculation of WoE and IV**

**Weight of Evidence- Woe**
To what extent an independent variable would predict a dependent variable

Positive WOE means Distribution of Goods > Distribution of Bads  
Negative WOE means Distribution of Goods < Distribution of Bads  
Hint : Log of a number > 1 means positive value. If less than 1, it means negative value.  

Two ways to classify groups in WoE calculations  

1. Fine Classing: Create 10 to 20 bins/groups for a continuous independent variable and then calculate WOE and IV of the variable. 
2. Coarse Classing: Combine adjacent categories with similar WOE scores

---
**Information Value (IV)**

Information value is a useful technique to select important variables in a predictive model. It represents how much independent information it brings originally to explain dependent value. It helps to rank variables on the basis of their importance. The IV is calculated using the following formula :  

IV = ∑ (% of non-events - % of events) * WOE

IV categories  
* Less than 0.02 => 	Not useful for prediction   
* 0.02 to 0.1	=> Weak predictive Power  
* 0.1 to 0.3	=> Medium predictive Power  
* 0.3 to 0.5 => 	Strong predictive Power  
* '>0.5 =>	Suspicious Predictive Power
